# Desafio

###1.
Para os residentes de Minas Gerais, qual categoria de produtos registra a média mais alta de valor total em suas compras?


```python
# Filtrar o DataFrame para incluir apenas os registros dos residentes de Minas Gerais
df_mg = df[df['estado'] == 'Minas Gerais']

# Calcular a média do valor total agrupada pela categoria de produto
media_por_categoria = df_mg.groupby('classe_produto')['valor_total_compra'].mean()

# Encontrar a categoria com a média mais alta
categoria_mais_alta = media_por_categoria.idxmax()

# Exibir a categoria com a média mais alta
print("Categoria com a média mais alta de valor total de compras em Minas Gerais:", categoria_mais_alta)
```

###2.
Qual produto registra o maior número de vendas entre consumidores diagnosticados com diabetes?


```python
df.head()
```

```python
# Filtrar o DataFrame para incluir apenas os registros dos consumidores diagnosticados com diabetes
df_diabetes = df[df['diabetes'] == 1]

# Calcular o número total de vendas de cada produto
vendas_por_produto = df_diabetes.groupby('nome_produto')['qtd_produto'].sum()

# Encontrar o produto com o maior número de vendas
produto_mais_vendido = vendas_por_produto.idxmax()

# Exibir o produto com o maior número de vendas
print("Produto com o maior número de vendas entre consumidores diagnosticados com diabetes:", produto_mais_vendido)
```

```python
# Filtrar o DataFrame para incluir apenas os registros dos consumidores diagnosticados com diabetes
df_diabetes = df[df['diabetes'] == 1]

# Calcular o número total de vendas de cada produto
vendas_por_produto = df_diabetes.groupby('nome_produto')['cod_cliente'].count()

# Encontrar o produto com o maior número de vendas
produto_mais_vendido = vendas_por_produto.idxmax()

# Exibir o produto com o maior número de vendas
print("Produto com o maior número de vendas entre consumidores diagnosticados com diabetes:", produto_mais_vendido)
```

###4.
Qual região registrou o maior valor total de compras de produtos?


```python
# Agrupar o DataFrame pela coluna 'regiao' e somar os valores totais de compra para cada região
total_por_regiao = df.groupby('regiao')['valor_total_compra'].sum()

# Encontrar a região com o maior valor total de compras
regiao_maior_valor = total_por_regiao.idxmax()

# Exibir a região com o maior valor total de compras
print("Região com o maior valor total de compras de produtos:", regiao_maior_valor)
```
